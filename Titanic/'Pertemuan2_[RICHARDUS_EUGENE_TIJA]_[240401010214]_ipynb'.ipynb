### Q2: Analisis Missing Values
Identifikasi kolom yang tidak lengkap datanya.

```python
missing_data = df.isnull().sum()
proportion = (missing_data / len(df)) * 100

summary_missing = pd.DataFrame({
    'Missing Values': missing_data,
    'Proporsi (%)': proportion.round(2)
})
print(summary_missing[summary_missing['Missing Values'] > 0])
```

### Q4: Filter Wanita Kelas 1
Analisis spesifik pada segmen penumpang tertentu.

```python
wanita_p1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
jumlah_wanita_p1 = len(wanita_p1)
rata_usia_wanita_p1 = wanita_p1['Age'].mean()

print(f'Jumlah wanita di Kelas 1: {jumlah_wanita_p1}')
print(f'Rata-rata usia mereka: {rata_usia_wanita_p1:.2f} tahun')
```

### Q5: Tingkat Keselamatan per Kelas
Membandingkan peluang selamat berdasarkan kelas tiket.

```python
safety_by_class = df.groupby('Pclass')['Survived'].mean().sort_values(ascending=False)
print(safety_by_class)
print(f'\nKelas dengan tingkat keselamatan tertinggi: Kelas {safety_by_class.idxmax()}')
```
